```python
import pandas as pd

filepath = 'salaries.csv'

def load_csv(filepath):
    data = []
    col = []
    checkcol = False
    with open(filepath) as f:
        for val in f.readlines():
            val = val.replace("\n", "")
            val = val.split(',')
            if not checkcol:
                col = val
                checkcol = True
            else:
                data.append(val)
    df = pd.DataFrame(data=data, columns=col)
    return df

loaded_df = load_csv(filepath)

print(loaded_df)
```
